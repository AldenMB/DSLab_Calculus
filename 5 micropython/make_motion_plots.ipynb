```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1)
yy = [
    x,
    1-(x>0.5)/2,
    (x-0.5)**2
]

for i, y in enumerate(yy):
    plt.plot(x, y)
    plt.axis('off')
    plt.savefig(f'exercise_8_{i}.png')
    plt.show()
```
