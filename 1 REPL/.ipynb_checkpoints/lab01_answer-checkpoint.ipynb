```python
x = np.linspace(0, math.pi, 50)
y = np.sin(100 * x)

# Graph for 5 points
plt.plot(x, y, 'r-')
plt.title("sin(100x)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

x = np.linspace(0, math.pi, 1000)
y = np.sin(100 * x)

# Graph for 5 points
plt.plot(x, y, 'r')
plt.title("sin(100x)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

__Part 3__

```python
plt.rcParams['figure.figsize'] = (10, 10)

x = np.linspace(-np.pi, np.pi, 120)

sin = np.sin(x)
d1 = x
d3 = d1 - x ** 3 / math.factorial(3)
d5 = d3 + x ** 5 / math.factorial(5)
d7 = d5 - x ** 7 / math.factorial(7)
d9 = d7 + x ** 9 / math.factorial(9)

plt.plot(x, sin, 'r-', x, d1, 'b-', x, d3, 'g-', x, d5, 'm-', x, d7, 'k-', x, d9, 'c-')
plt.legend(['sin(x)', 'd1', 'd3', 'd5', 'd7', 'd9'], loc='lower right')
plt.title("sin(x) Taylor Approximations at various orders of approximation")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
